 2. What is an interpreter in Python?
An interpreter is a program that executes code line by line, converting Python code into machine code. Python's interpreter processes the code during runtime, making it easy to debug but slower than compiled languages like C++.

 3. What are pre-defined keywords in Python?
Keywords are reserved words in Python that have special meanings — they can't be used for anything else like variable names or functions.

 Examples: if, else, for, while, def, import, try, except, class, True, False, None

 4. Can keywords be used as variable names?
No — Python keywords cannot be used as variable names. Doing so throws a SyntaxError.

5. What is mutability in Python?
Mutability refers to whether an object can be changed after creation.

Mutable: Can be modified (e.g., list, dict, set)

Immutable: Cannot be modified (e.g., int, str, tuple)

 6. Why are lists mutable, but tuples are immutable?
Lists: Designed for dynamic data — elements can be added, removed, or modified.

Tuples: Designed for fixed data — they’re faster and safer from accidental changes.

 7. What is the difference between "==" and "is" operators in Python?
== checks if values are equal.

is checks if two objects point to the same memory location (identity).


8. What are logical operators in Python?
Logical operators combine conditional statements:

and — returns True if both statements are true

or — returns True if one of the statements is true

not — reverses the result (true becomes false)

 9. What is type casting in Python?
Type casting (or type conversion) converts data from one type to another.

 10. What is the difference between implicit and explicit type casting?
Implicit type casting: Python automatically converts types where needed.

a = 5       # int
b = 2.5     # float
c = a + b   # c becomes float (7.5)
Explicit type casting: Done manually using functions like int(), float(), str(), etc.

11. What is the purpose of conditional statements in Python?
Conditional statements control the flow of your program based on certain conditions.

12. How does the elif statement work?
The elif statement (short for else if) lets you check multiple conditions after an if statement. It’s useful when you want different outcomes based on different conditions.

13.What is the difference between for and while loops?
for loop : iterates over asequence , use for item in sequence
while loop : Repeats as long as a condition is True , Uses while condition

14.Describe a scenario where a while loop is more suitable than a for loop.

Scenario: Imagine you’re creating a program that keeps asking a user for a correct password until they get it right. Since you don’t know how many attempts it will take, a while loop is better.
Here’s why while is better:

The loop runs indefinitely until the correct input is provided.

A for loop wouldn’t work well because we don’t know how many times the user will guess.
